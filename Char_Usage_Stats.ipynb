```python
import cPickle
import numpy as np
import seaborn as sns
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

from collections import Counter
```

```python
with open('data/char_usage_accum.p','rb') as f:
     all_chars_accum = cPickle.load(f)
```

```python
with open('data/player_to_country.p','rb') as f:
     player_to_country = cPickle.load(f)
```

```python
countries_data = [player_to_country[i] for i in player_to_country if player_to_country[i] != '']
```

```python
country_to_count = Counter(countries_data)
```

```python
d = pd.Series(country_to_count)
```

```python
fig,ax = plt.subplots(figsize=(16,12))
x = d.sort_values(ascending=False)
y = d.sort_values(ascending=False).index
sns.barplot(x, y, palette="Greens_d");
```

```python
char_to_count = Counter(all_chars_accum)
char_to_count['Dan']=0
s = pd.Series(char_to_count)
```

```python
fig,ax = plt.subplots(figsize=(16,12))
x = s.sort_values(ascending=False)
y = s.sort_values(ascending=False).index
sns.barplot(x, y, palette="Greens_d");
```
